## Part 2

```python
# Initialize computer network
network = []
for a in range(50):
    name = "NIC"+str(a)
    c = IntcodeV3(prog,name)
    c.setInputs([a]) # initialize with netwrok address
    c.runProgram()
    network.append(c)
    
# Initialize instruction queues 
from queue import Queue
queues = []
for a in range(50):
    q = Queue()
    q.put([-1]) # initialize with empty instruction
    queues.append(q)

# Initialize NAT
natq = [-1]
nidle = 0
yinst = []

# Run network loop    
ended = False
inst = []

istep = 0

while(True):
    
    istep += 1
    
    # one network cycle
    for a in range(50):
        #print("Running computer # ",a)
        c = network[a]
        # get next instruction
        q = queues[a]
        i = [-1]
        if q.empty():
            q.put(i)
        else:
            i = q.get()
        # run computer with current instruction
        # print(i)
        c.setInputs(i)
        c.runProgram()
        o = c.getOutput()
        c.resetOutput()
        # parse the output, if any
        if len(o):
            for x in range(0,len(o),3):
                addr = o[x]
                inst = o[x+1:x+3]
                if addr==255:
                    # print(istep, "--> Sending instruction",inst,"to NAT")
                    natq = inst
                else:
                    queues[addr].put(inst)     
    # NAT cycle
    idle = 0
    for a in range(50):
        q = queues[a]
        if q.empty():
            idle +=1
        elif q.qsize() == 1: # only one instruction in queue
            i = q.get()
            if len(i)==1: # instruction have lenght 1, thus is [-1]
                idle +=1
            q.put(i) # put back instruction      
    if idle==50: # all computers have idle queues, send NAT instruction to computer 0
        print("Network is idle, sending instruction",natq,"to computer 0")
        q = queues[0]
        if q.qsize()==1:
            i = q.get() # remove last [-1] from queue
        q.put(natq) # put last NAT instruction
        nidle += 1
        if natq[1] in yinst:
            print("Dound duplicate Y instruction:",natq[1])
            break
        yinst.append(natq[1])
```
