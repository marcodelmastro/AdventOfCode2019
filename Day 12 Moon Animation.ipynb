# Day 12 Moon Animation

```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

from celluloid import Camera

from AOC2019 import MoonSystem
```

```python
# Input
x1 = np.array([7, 10, 17])
x2 = np.array([-2, 7, 0])
x3 = np.array([12, 5, 12])
x4 = np.array([5, -8, 6])

s = MoonSystem([x1,x2,x3,x4])

istep = 0
istepmax = 50 # number of steps, will be number of frames in animation

npoints = len(s.x)
colors = cm.rainbow(np.linspace(0,1,npoints))

fig = plt.figure()
ax = Axes3D(fig) # for 3D scatter plot
camera = Camera(fig)

print("Generating coordinates and scatter plots...")

while(istep<=istepmax):    
    x = []
    y = []
    z = []
    zz = []
    for i in range(npoints):
        x.append(s.x[i][0])
        y.append(s.x[i][1])
        z.append(s.x[i][2])
        zz.append(10*s.x[i][2]+100)
    # 2D
    #points = [x,y]
    #plt.scatter(*points,c=colors,s=zz)
    # 3D
    points = [x,y,z]
    ax.scatter(*points,c=colors,s=200)
    camera.snap()
    s.updateSystem()
    istep += 1

print("Animating scatter plots...")
anim = camera.animate(blit=True)

print("Saving animation...",end="")
#anim.save('./visualisation/day12animation.mp4',fps=10)
anim.save('./visualisation/day12animation.gif', writer='imagemagick', fps=10)
print(" done.")
```
